```python
import seaborn as sns
sns.set_style('darkgrid')
import matplotlib.pyplot as plt
```

```python
fig, (ax1,ax2) = plt.subplots(2,1, sharex=True)
ax1 = sns.barplot(x=[0.2,0.4,0.6,0.8], y=[0.1948,0.00837,0.0001314,0.00007421],ax=ax1, palette="Blues_d")
ax1.set(yscale="log")
ax1.set_ylabel("Mean Squared Error (MAE)")
# ax1.set_xlabel("kappa")

ax2 = sns.barplot(x=[0.2,0.4,0.6,0.8], y=[2.3,2.1,2.3,2.3],ax=ax2,palette="Blues_d")
ax2.set_ylabel("Compression Ratio")
ax2.set_xlabel("kappa")
ax1.set_title("Network Test Case 14s")
plt.tight_layout()
plt.savefig("kappa_14s.pdf")
```
