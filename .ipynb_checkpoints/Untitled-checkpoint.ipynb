```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.set_style("darkgrid")
```

```python
fig, ax = plt.subplots(2,1)
ax[0] = sns.barplot(x=[0.2,0.4,0.6,0.8], y=[0.00047074,0.00042891,0.00012740,0.00004434],palette="Blues_r", ax=ax[0])
ax[0].set_ylabel('Mean Squared Error (MSE)')
ax[0].set(xticklabels=[])
ax[1] = sns.barplot(x=[0.2,0.4,0.6,0.8], y=[2.0,2.2,2.2,2.2],palette="Blues_r", ax=ax[1])
ax[1].set_ylabel('Compression Ratio')
ax[1].set_xlabel("kappa")
ax[0].set_title("Case opf-storage-hvdc")
#plt.tight_layout()
plt.show()
plt.savefig("kappa_case_opf.pdf")
plt.close()
```

```python
fig, ax = plt.subplots(2,1)
ax[0] = sns.barplot(x=['(5,5)','(10,5)','(15,10)','(20,10)'], y=[0.00075461, 0.00042891,0.00009454,0.00002809],palette="Greens_r", ax=ax[0])
ax[0].set_ylabel('Mean Squared Error (MSE)')
ax[0].set(xticklabels=[])
ax[1] = sns.barplot(x=['(5,5)','(10,5)','(15,10)','(20,10)'], y=[2.1,2.2,2.0,2.0],palette="Greens_r", ax=ax[1])
ax[1].set_ylabel('Compression Ratio')
ax[1].set_xlabel("Layer Multiplying Factors")
ax[0].set_title("Case opf-storage-hvdc")
plt.ticklabel_format(axis="y", style="scientific")
#plt.tight_layout()
plt.show()
plt.savefig("lmf_case_opf.pdf")
plt.close()
```
